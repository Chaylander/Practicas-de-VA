```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps

Letra = "R"

contador=0

while contador<11:
    opcion = input("Escribe la inicial de mi nombre ")
    #print(opcion)
    
    if opcion == "R":
        
        contador=contador+1
        print (contador)
        
        if contador != 12:
            img1 = cv2.imread('monokuma_res.jpg', cv2.IMREAD_GRAYSCALE)
            cv2.imshow('Imagen 1',img1)
            img2 = cv2.imread('thewitcher_res.jpg', cv2.IMREAD_GRAYSCALE)
            cv2.imshow('Imagen 2',img2)
            cv2.moveWindow('Imagen 1', -16,0)
            cv2.moveWindow('Imagen 2', 1325,0)
            #ecualizada imagen 1
            ecu1 = cv2.equalizeHist(img1)
            cv2.imshow('imagen ecualizada', ecu1) 
            cv2.moveWindow('imagen ecualizada', -16,500)
            #ecualizado imagen 2
            ecu2 = cv2.equalizeHist(img2)
            cv2.imshow('imagen ecualizada 2', ecu2) 
            cv2.moveWindow('imagen ecualizada 2', 1325,500)
            #Histograma imagen 1
            print("Histograma imagen 1")
            plt.hist(img1.ravel())
            plt.xlabel('intensidad de iluminacion')
            plt.ylabel('cantidad de pixeles')
            plt.title('Histograma de la Imagen 1')
            plt.show()
            #Histograma imagen 2
            print("Histograma imagen 2")
            plt.hist(img2.ravel())
            plt.xlabel('intensidad de iluminacion')
            plt.ylabel('cantidad de pixeles')
            plt.title('Histograma de la Imagen 2')
            plt.show()
            #Histograma imagen ecualizada 1
            print("Histograma ecualizado img1")
            plt.hist(ecu1.ravel())
            plt.xlabel('intensidad de iluminacion')
            plt.ylabel('cantidad de pixeles')
            plt.title('Ecualizacion del histograma de la img 1')
            plt.show()
            #Histograma imagen ecualizada 2
            print("Histograma ecualizado img2")
            plt.hist(ecu2.ravel())
            plt.xlabel('intensidad de iluminacion')
            plt.ylabel('cantidad de pixeles')
            plt.title('Ecualizacion del histograma de la img  2')
            plt.show()
            
        
            
            if contador == 1:
                print("Suma de imagenes------------------------------------------------------------")
                ressuma = cv2.addWeighted(img1,0.5,img2,0.5,1)
                cv2.imshow('suma',ressuma)
                cv2.moveWindow('suma', 650,0)
                
                #Imagen ecualizada de la suma
                ecusuma = cv2.equalizeHist(ressuma)
                cv2.imshow('imagen suma ecualizada', ecusuma) 
                cv2.moveWindow('imagen suma ecualizada', 650,500)
                
                #Histograma de la suma
                print("Histograma de la suma")
                plt.hist(ressuma.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('histograma de la suma ')
                plt.show()
                
                #Histograma de la suma ecualizada
                print("Histograma de la suma ecualizada")
                plt.hist(ecusuma.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('Ecualizacion del histograma de la suma')
                plt.show()

                              
                cv2.waitKey(0)
                cv2.destroyAllWindows()
        
            elif contador == 2:
                print("Resta de imagenes-----------------------------------------------------------")
                resresta = cv2.absdiff(img1,img2)
                
                
                #print('img1[0,0]= ', img1[0,0])
                #print('img2[0,0]= ',img2[0,0])
                #print('resultado[0,0]= ',resresta[0,0])
                            
                cv2.imshow('resta',resresta)
                cv2.moveWindow('resta', 650,0)
                
                #Imagen ecualizada de la resta
                ecuresta = cv2.equalizeHist(resresta)
                cv2.imshow('imagen resta ecualizada', ecuresta) 
                cv2.moveWindow('imagen resta ecualizada', 650,500)
                
                #Histograma de la resta
                print("Histograma de la resta")
                plt.hist(resresta.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('Histograma de la resta')
                plt.show()
                
                #Histograma de la resta ecualizada
                print("Histograma de la resta ecualizada")
                plt.hist(ecuresta.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('Ecualizacion del histograma de la resta')
                plt.show()
                
                cv2.imwrite('resta.jpg',resresta)
                cv2.waitKey(0)
                cv2.destroyAllWindows()
            
            elif contador == 3:
                print("Division--------------------------------------------------------------------")
                
                #resdiv = cv2.divide(img1,0.5,img2,0.1,1)
                resdiv=cv2.divide(img1,img2)
                
                #print('img1[0,0]= ', img1[0,0])
                #print('img2[0,0]= ',img2[0,0])
                #print('resultado[0,0]= ',resdiv[0,0])
                
                cv2.imshow('division',resdiv)
                cv2.moveWindow('division', 650,0)
                cv2.imwrite('division.jpg',resdiv)
                
                #Imagen ecualizada de la division
                ecudiv = cv2.equalizeHist(resdiv)
                cv2.imshow('imagen division ecualizada', ecudiv) 
                cv2.moveWindow('imagen division ecualizada', 650,500)
                
                #Histograma de la div
                print("Histograma de la division")
                plt.hist(resdiv.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('histograma de la division')
                plt.show()
                
                #Histograma de la div ecualizada
                print("Histograma de la division ecualizada")
                plt.hist(ecudiv.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('Ecualizacion del histograma de la division')
                plt.show()
                
                cv2.waitKey(0)
                cv2.destroyAllWindows()
            
            elif contador == 4:
                print("Multiplicacion---------------------------------------------------------------")
            
                resmul=cv2.multiply(img1,img2)
                
                #print('img1[0,0]= ', img1[0,0])
                #print('img2[0,0]= ',img2[0,0])
                #print('resultado[0,0]= ',resmul[0,0])
                
                cv2.imshow('multiplicacion',resmul)
                cv2.moveWindow('multiplicacion', 650,0)
                
                #Imagen ecualizada de la multiplicacion
                ecumul = cv2.equalizeHist(resmul)
                cv2.imshow('imagen multiplicacion ecualizada', ecudiv) 
                cv2.moveWindow('imagen multiplicacion ecualizada', 650,500)
                
                #Histograma de la multiplicacion
                print("Histograma de la multiplicacion")
                plt.hist(resmul.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('histograma de la multiplicacion')
                plt.show()
                
                #Histograma de la multiplicacion ecualizada
                print("Histograma de la multiplicacion ecualizada")
                plt.hist(ecumul.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('Ecualizacion del histograma de la multiplicacion')
                plt.show()
                
                cv2.imwrite('multiplicacion.jpg',resmul)
                cv2.waitKey(0)
                cv2.destroyAllWindows()
            
            elif contador == 5:
                print("Conjuncion-------------------------------------------------------------------")
            
                resconjuncion=cv2.bitwise_and (img1, img2)
                
                #print('img1[0,0]= ', img1[0,0])
                #print('img2[0,0]= ',img2[0,0])
                #print('resconjuncion[0,0]= ',resmul[0,0])
                
                cv2.imshow('Conjuncion',resconjuncion)
                cv2.moveWindow('Conjuncion', 650,0)
                
                #Imagen ecualizada de la conjuncion
                ecuconj = cv2.equalizeHist(resconjuncion)
                cv2.imshow('imagen conjuncion ecualizada', ecuconj) 
                cv2.moveWindow('imagen conjuncion ecualizada', 650,500)
                
                #Histograma de la conjuncion
                print("Histograma de la conjuncion")
                plt.hist(resconjuncion.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('histograma de la conjuncion')
                plt.show()
                
                #Histograma de la conjuncion ecualizada
                print("Histograma de la conjuncion ecualizada")
                plt.hist(ecuconj.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('Ecualizacion del histograma de la conjuncion') 
                plt.show()
                
                cv2.imwrite('Conjuncion.jpg',resconjuncion)
                
                cv2.waitKey(0)
                cv2.destroyAllWindows()
            
            elif contador == 6:
                print("Disyuncion-----------------------------------------------------------------")
    
                resdisyuncion=cv2.bitwise_or(img1, img2)
                
                #print('img1[0,0]= ', img1[0,0])
                #print('img2[0,0]= ',img2[0,0])
                #print('resultado[0,0]= ',resmul[0,0])
                
                cv2.imshow('disyuncion',resdisyuncion)
                cv2.moveWindow('disyuncion', 650,0)
                
                #Imagen ecualizada de la disyuncion
                ecudis = cv2.equalizeHist(resdisyuncion)
                cv2.imshow('imagen disyuncion ecualizada', ecudis) 
                cv2.moveWindow('imagen disyuncion ecualizada', 650,500)
                
                #Histograma de la disyuncion
                print("Histograma de la disyuncion")
                plt.hist(resdisyuncion.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('histograma de la disyuncion')
                plt.show()
                
                #Histograma de la disyuncion ecualizada
                print("Histograma de la disyuncion ecualizada")
                plt.hist(ecudis.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('Ecualizacion del histograma de la disyuncion')
                plt.show()
                
                cv2.imwrite('disyuncion.jpg',resdisyuncion)
                cv2.waitKey(0)
                cv2.destroyAllWindows()
            
            elif contador == 7:
                print("Negacion---------------------------------------------------------------------")
    
                resnegacion=cv2.bitwise_not(img1)
                
                #print('img1[0,0]= ', img1[0,0])
                #print('resnegacion[0,0]= ',resnegacion[0,0])
                
                cv2.imshow('negacion',resnegacion)
                
                #Imagen ecualizada de la negacion
                ecuneg = cv2.equalizeHist(resnegacion)
                cv2.imshow('imagen negacion ecualizada', ecuneg) 
                cv2.moveWindow('imagen negacion ecualizada', 650,500)
                
                #Histograma de la negacion
                print("Histograma de la negacion")
                plt.hist(resnegacion.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('histograma de la negacion')
                plt.show()
                
                #Histograma de la negacion ecualizada
                print("Histograma de la negacion ecualizada")
                plt.hist(ecuneg.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('Ecualizacion del histograma de la negacion')
                plt.show()
                
                
                cv2.imwrite('negacion.jpg',resnegacion)
                cv2.moveWindow('negacion', 650,0)
                cv2.waitKey(0)
                cv2.destroyAllWindows()
            
            elif contador == 8:
                print("Traslacion------------------------------------------------------------------")
    
                img = cv2.imread('monokuma_res.jpg',0)
                rows,cols = img.shape
                
                M = np.float32([[1,0,210],[0,1,20]])
                dst = cv2.warpAffine(img,M,(cols,rows))
                
                cv2.imshow('img',dst)
                cv2.moveWindow('img', 650,0)
                
                #Imagen ecualizada de la traslacion
                ecutrasla = cv2.equalizeHist(dst)
                cv2.imshow('imagen traslacion ecualizada', ecutrasla) 
                cv2.moveWindow('imagen traslacion ecualizada', 650,500)
                
                #Histograma de la traslacion
                print("Histograma de la traslacion")
                plt.hist(dst.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('histograma de la traslacion')
                plt.show()
                
                #Histograma de la rotacion ecualizada
                print("Histograma de la traslacion ecualizada")
                plt.hist(ecutrasla.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('Ecualizacion del histograma de la traslacion')
                plt.show()
                
                cv2.waitKey(0)
                cv2.destroyAllWindows()
            
            elif contador == 9:
                print("Escalado")
                img = cv2.imread("monokuma_res.jpg", cv2.IMREAD_GRAYSCALE)
                newImg = cv2.resize(img, (0,0), fx=0.3, fy=0.3)
                cv2.imwrite('escalado.jpg',newImg)
                cv2.imshow('Resized Cuadrito1', newImg)
                cv2.moveWindow('Resized Cuadrito1', 650,0)
                
                 
                #Imagen ecualizada del escalado
                ecuescalado = cv2.equalizeHist(newImg)
                cv2.imshow('imagen escalada ecualizada', ecuescalado) 
                cv2.moveWindow('imagen escalada ecualizada', 650,500)
                
                #Histograma del escalado
                print("Histograma del escalado")
                plt.hist(newImg.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('histograma del escalado')
                plt.show()
                
                #Histograma del escalado ecualizada
                print("Histograma del escalado ecualizada")
                plt.hist(ecuescalado.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('Ecualizacion del histograma del escalado')
                plt.show()
                
                cv2.waitKey(0)
                cv2.destroyAllWindows()
                
            
            elif contador ==10:
                print("Rotacion-------------------------------------------------------------------")
                img = cv2.imread('monokuma_res.jpg',0)
                rows,cols = img.shape
                
                M = cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
                dst = cv2.warpAffine(img,M,(cols,rows))
                cv2.imshow('Rotacion', dst)
                cv2.moveWindow('Rotacion', 650,0)
                
                #Imagen ecualizada de la rotacion
                ecurota = cv2.equalizeHist(dst)
                cv2.imshow('imagen rotacion ecualizada', ecurota) 
                cv2.moveWindow('imagen rotacion ecualizada', 650,500)
                
                #Histograma de la rotacion
                print("Histograma de la rotacion")
                plt.hist(dst.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('histograma de la rotacion')
                plt.show()
                
                #Histograma de la rotacion ecualizada
                print("Histograma de la rotacion ecualizada")
                plt.hist(ecurota.ravel())
                plt.xlabel('intensidad de iluminacion')
                plt.ylabel('cantidad de pixeles')
                plt.title('Ecualizacion del histograma de la rotacion')
                plt.show()
                
                cv2.waitKey(0)
                cv2.destroyAllWindows()
                
            else :
                print('Traslacion a fin----------------------------------------------------------')
                
                
                img = cv2.imread('monokuma_res.jpg')
                rows,cols,ch = img.shape
            
    
                pts1 = np.float32([[100,400],[400,100],[100,100]])
                pts2 = np.float32([[50,300],[400,200],[80,150]])
    
    
                M = cv2.getAffineTransform(pts1,pts2)
                dst = cv2.warpAffine(img,M,(cols,rows))
    
                plt.subplot(121),plt.imshow(img),plt.title('Entrada')
                plt.subplot(122),plt.imshow(dst),plt.title('Salida')
                plt.show()
            
print("Ha terminado")
```
