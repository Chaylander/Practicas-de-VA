Impresion en blanco y negro con cv2

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

imagen = cv2.imread('desktop1.jpg',cv2.IMREAD_GRAYSCALE)

cv2.imshow('Cuadrito1',imagen)

cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname = 'desktop1.jpg'
image = Image.open(fname).convert("L")
arr = np.asarray(image)
plt.imshow(arr, cmap='gray', vmin=0, vmax=255)
plt.show()
```

Impresi√≥n cambiando la resoluci√≥n espacial

```python
import cv2

img = cv2.imread("desktop1.jpg",cv2.IMREAD_GRAYSCALE)

newImg = cv2.resize(img, (0,0), fx=0.3, fy=0.3)

cv2.imshow('Resized Cuadrito1', newImg)

cv2.waitKey(0)
```
